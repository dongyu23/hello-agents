## 2. 基础功能演示

### 2.1 初始化 LLM 客户端

```python
from app.core.llm_client import LLMClient

# 初始化客户端
llm_client = LLMClient(
    api_key=os.getenv("API_KEY"),
    model_name=os.getenv("MODEL_NAME", "glm-4.5"),
    base_url=os.getenv("BASE_URL", "https://open.bigmodel.cn/api/paas/v4/")
)

print("✅ LLM 客户端初始化完成")
```

### 2.2 测试基础对话

```python
async def test_basic_chat():
    response = await llm_client.chat_completion(
        messages=[
            {"role": "system", "content": "你是一个有帮助的助手。"},
            {"role": "user", "content": "你好，请用一句话介绍你自己。"}
        ]
    )
    return response

# 运行测试
result = asyncio.run(test_basic_chat())
print(f"助手回复: {result}")
```

## 3. 创建多智能体讨论

### 3.1 定义讨论参与者

```python
# 定义参与讨论的角色
participants = [
    "苏格拉底",
    "埃隆·马斯克",
    "孔子"
]

# 定义讨论主题
topic = "人工智能是否会改变人类的本质？"

print(f"讨论主题: {topic}")
print(f"参与角色: {', '.join(participants)}")
```

### 3.2 创建论坛

```python
from app.services.forum_service import ForumService
from app.models.forum import ForumCreate

async def create_forum():
    forum_service = ForumService()
    
    forum_data = ForumCreate(
        title="AI时代的哲学思考",
        topic=topic,
        participants=participants
    )
    
    forum = await forum_service.create_forum(forum_data)
    return forum

# 创建论坛
forum = asyncio.run(create_forum())
print(f"✅ 论坛创建成功")
print(f"论坛ID: {forum.id}")
print(f"标题: {forum.title}")
```

### 3.3 模拟讨论流程

```python
async def simulate_discussion(forum_id: str, rounds: int = 3):
    """模拟多轮讨论"""
    from app.services.forum_scheduler import ForumScheduler
    
    scheduler = ForumScheduler()
    
    print("\n" + "="*50)
    print("🎭 开始多智能体圆桌讨论")
    print("="*50 + "\n")
    
    # 启动论坛
    await scheduler.start_forum(forum_id)
    
    # 模拟多轮对话
    for i in range(rounds):
        print(f"\n📢 第 {i+1} 轮讨论")
        print("-" * 30)
        
        # 获取当前发言
        messages = await scheduler.get_forum_messages(forum_id)
        
        for msg in messages[-len(participants):]:
            print(f"\n{msg['speaker']}: {msg['content'][:200]}..." if len(msg['content']) > 200 else f"\n{msg['speaker']}: {msg['content']}")
    
    print("\n" + "="*50)
    print("✅ 讨论完成")
    print("="*50)

# 运行讨论模拟
# asyncio.run(simulate_discussion(forum.id, rounds=3))
print("(此功能需要完整的服务器环境，请运行完整项目体验)')
```

## 4. 角色生成与定制

### 4.1 使用 RealGod Agent 生成角色

```python
from app.agents.god_agent import GodAgent

async def generate_persona(name: str):
    """生成角色人设"""
    god_agent = GodAgent(llm_client=llm_client)
    
    print(f"\n正在生成 {name} 的角色设定...")
    persona = await god_agent.generate_persona(name)
    
    print(f"\n✅ {name} 角色生成完成")
    print(f"姓名: {persona.name}")
    print(f"性格: {persona.personality}")
    print(f"说话风格: {persona.speaking_style}")
    
    return persona

# 生成角色
# socrates = asyncio.run(generate_persona("苏格拉底"))
# musk = asyncio.run(generate_persona("埃隆·马斯克"))
print("(此功能需要完整的服务器环境，请运行完整项目体验)')
```

## 5. 评估讨论质量

### 5.1 五维评估指标

```python
from app.evaluation.metrics import DiscussionMetrics

def explain_metrics():
    """解释评估指标"""
    metrics = {
        "观点多样性": "衡量不同智能体提出的观点的差异化程度",
        "深度演进": "评估讨论过程中思想的深入和演进程度",
        "交互批判性": "分析智能体之间的质疑、反驳和相互启发",
        "连贯性": "评估讨论的逻辑连贯性和主题一致性",
        "创新性": "衡量讨论中产生的新观点和创意想法"
    }
    
    print("📊 MADF 五维评估指标")
    print("="*50)
    for name, description in metrics.items():
        print(f"\n{name}:")
        print(f"  {description}")

explain_metrics()
```
